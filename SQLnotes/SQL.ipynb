```python
#Aggregating information at "Embarked" and "Sex" levels
ps.sqldf("""SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df
GROUP BY df.Embarked,df.Sex """)
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
ps.sqldf(""" SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
FROM titanic_disaster_dataset AS df
WHERE age >= 18 AND age <=60    
GROUP BY df.Embarked,df.Sex
""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output
ps.sqldf("""SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df
WHERE age >= 18 AND age <=60   
GROUP BY df.Embarked,df.Sex
ORDER BY df.Embarked,df.Sex
""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output in descending order
ps.sqldf("""SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df
WHERE age >= 18 AND age <=60 
GROUP BY df.Embarked,df.Sex
ORDER BY df.Embarked DESC,df.Sex DESC
""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also using "HAVING" for fitering the output for buckets where average age is more than 30
#and also using "ORDER BY" for sorting the output in descending order
ps.sqldf("""SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df

WHERE age >= 18 AND age <=60
GROUP BY df.Embarked,df.Sex
HAVING avg_age > 30
ORDER BY df.Embarked DESC,df.Sex DESC
""")
```

```python
#RANK() sorts the data, gives the same rank to all consecutive n rows with same values, 
#but for next row with a different value, it gives a rank equal to previous rank plus n
ps.sqldf("""
SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(ORDER BY df.Age DESC) AS rnk
FROM titanic_disaster_dataset AS df
""")
```

```python
#DENSE_RANK() does not skip ranks like RANK() does
ps.sqldf("""
SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    DENSE_RANK() OVER(ORDER BY df.Age DESC) AS rnk
    
FROM titanic_disaster_dataset AS df
""")
```

```python
#ROW_NUMBER() sorts the data and gives different ranks to each row 
ps.sqldf("""
SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    ROW_NUMBER() OVER(ORDER BY df.Age DESC) AS rnk
    
FROM titanic_disaster_dataset AS df
""")
```

```python
#You can use multiple columns to sort the data within OVER() function
ps.sqldf("""
SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(ORDER BY df.Age DESC, df.Fare ASC) AS rnk
    
FROM titanic_disaster_dataset AS df
""")
```

```python
#Using PARTITION BY to create buckets for giving separate sets of ranks
ps.sqldf("""
SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(PARTITION BY df.Embarked, df.Sex 
                ORDER BY df.Age DESC, df.Fare DESC) AS rnk
    
FROM titanic_disaster_dataset AS df

""")
```

```python
#Examples for LEAD() and LAG()
ps.sqldf("""
SELECT 
    df.Name,
    LEAD(df.Name, 1) OVER(ORDER BY df.Name) AS next_name,
    LEAD(df.Name, 2) OVER(ORDER BY df.Name) AS second_next_name,
    LAG(df.Name, 1) OVER(ORDER BY df.Name) AS previous_name,
    LAG(df.Name, 2) OVER(ORDER BY df.Name) AS second_previous_name
    
FROM titanic_disaster_dataset AS df
""")
```

```python
#Using subqueries to get average age and then filtering only those rows where average age is below 30
ps.sqldf("""
SELECT 

    df1.Embarked,
    df1.avg_age
    
FROM (SELECT 
            a.Embarked, 
            AVG(a.Age) AS avg_age 
        FROM titanic_disaster_dataset AS a
        GROUP BY a.Embarked) AS df1

WHERE df1.avg_age<30
    
""")
```

```python
#Using subqueries to get average age and number of passengers separately and then joining the outputs
ps.sqldf("""
SELECT 
    df1.Embarked,
    df1.avg_age,
    df2.tot_passengers
FROM (SELECT 
            a.Embarked, 
            AVG(a.Age) AS avg_age 
        FROM titanic_disaster_dataset AS a
        GROUP BY a.Embarked) AS df1

INNER JOIN (SELECT 
                b.Embarked, 
                COUNT(b.PassengerID) AS tot_passengers
            FROM titanic_disaster_dataset AS b
            GROUP BY b.Embarked) AS df2
    ON df1.Embarked = df2.Embarked
    
""")
```

```python
#Using nested subqueries to filter data
ps.sqldf("""
SELECT b.* FROM titanic_disaster_dataset AS b
WHERE b.Embarked IN 
       (SELECT 
            DISTINCT df1.Embarked
        FROM (SELECT 
                    a.Embarked, 
                    AVG(a.Age) AS avg_age 
                FROM titanic_disaster_dataset AS a
                GROUP BY a.Embarked) AS df1
        WHERE df1.avg_age>30)
    
""")
```

```python
#Using CAST() to change datatype of columns
ps.sqldf("""
SELECT 
    df.Name,
    df.Fare,
    CAST(df.Fare AS INT) AS integer_fare,
    CAST(CAST(df.Fare AS INT) AS FLOAT) AS float_fare
FROM titanic_disaster_dataset AS df
LIMIT 5
    
""")
```

```python
#Samples
#ROWS BETWEEN clause is used to define a window frame for analytic functions, specifying which rows are included in the window of computation.
```

```python
#This query returns the last sale amount for each product across the entire partition of rows for that product.
#Using UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
SELECT
    product_id,
    sale_date,
    amount,
    LAST_VALUE(amount) OVER (
        PARTITION BY product_id 
        ORDER BY sale_date 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
    ) AS last_sale_amount
FROM
    sales;
```

```python
#This query returns the last sale amount considering only the current row and the previous row for each product. 
#This can help to see the change in the last sale amount dynamically as you move through the rows.
#Using ROWS 1 PRECEDING
SELECT
    product_id,
    sale_date,
    amount,
    LAST_VALUE(amount) OVER (
        PARTITION BY product_id 
        ORDER BY sale_date 
        ROWS BETWEEN 1 PRECEDING AND CURRENT ROW
    ) AS last_sale_amount
FROM
    sales;
```

```python
SELECT
    product_id,
    sale_date,
    amount,
    LAST_VALUE(amount) OVER (PARTITION BY product_id ORDER BY sale_date ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS last_sale_amount
FROM
    sales;
```

```python
SELECT
    product_id,
    sale_date,
    amount,
    FIRST_VALUE(amount) OVER (PARTITION BY product_id ORDER BY sale_date) AS first_sale_amount
FROM
    sales;
```

```python
#NTH_VALUE Analytic Function

SELECT empno,
       ename,
       deptno,
       sal,
       NTH_VALUE(sal, 3) FROM FIRST OVER (ORDER BY sal ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS third_lowest_sal,
       NTH_VALUE(sal, 3) FROM LAST OVER (ORDER BY sal ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS third_highest_sal
FROM   emp;
```

```python
SELECT empno,
       ename,
       deptno,
       sal,
       NTH_VALUE(sal, 2) FROM FIRST OVER (PARTITION BY deptno ORDER BY sal
         ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS second_lowest_sal,
       NTH_VALUE(sal, 2) FROM LAST OVER (PARTITION BY deptno ORDER BY sal 
         ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS second_highest_sal
FROM   emp;
```

```python
#PERCENT_RANK Analytic Function

SELECT empno,
       ename,
       deptno,
       sal,
       PERCENT_RANK() OVER (ORDER BY sal) AS percent_rank_sal,
       ROUND(PERCENT_RANK() OVER (ORDER BY sal)*100) AS percent_rank_sal_pct
FROM   emp;
```
